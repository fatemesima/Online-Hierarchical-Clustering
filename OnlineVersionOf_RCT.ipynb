```python
import math
def dist(x,y):
    dis = 0
    d = len(x)
    for i in range(d):
        dis += (x[i]-y[i])**2
    return math.sqrt(dis)


def diameter(s):
    diam = 0
    for x in s:
        for y in s:
            dis = dist(x,y)
            if diam < dis:
                diam = dis
    return diam
```

```python
class Node:
    def __init__(self, val):
        # self.parent = None
        self.left = None
        self.right = None
        self.element = val

    def __iter__(self):
        if self.left:
            yield from self.left

        yield self.element

        if self.right:
            yield from self.right

        if self.parent:
          yield from self.parent



    def display(self):
            lines, *_ = self._display_aux()
            for line in lines:
                print(line)

    def _display_aux(self):
            """Returns list of strings, width, height, and horizontal coordinate of the root."""
            # No child.
            if self.right is None and self.left is None:
                line = '%s' % self.element
                width = len(line)
                height = 1
                middle = width // 2
                return [line], width, height, middle

            # Only left child.
            if self.right is None:
                lines, n, p, x = self.left._display_aux()
                s = '%s' % self.element
                u = len(s)
                first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s
                second_line = x * ' ' + '/' + (n - x - 1 + u) * ' '
                shifted_lines = [line + u * ' ' for line in lines]
                return [first_line, second_line] + shifted_lines, n + u, p + 2, n + u // 2

            # Only right child.
            if self.left is None:
                lines, n, p, x = self.right._display_aux()
                s = '%s' % self.element
                u = len(s)
                first_line = s + x * '_' + (n - x) * ' '
                second_line = (u + x) * ' ' + '\\' + (n - x - 1) * ' '
                shifted_lines = [u * ' ' + line for line in lines]
                return [first_line, second_line] + shifted_lines, n + u, p + 2, u // 2

            # Two children.
            left, n, p, x = self.left._display_aux()
            right, m, q, y = self.right._display_aux()
            s = '%s' % self.element
            u = len(s)
            first_line = (x + 1) * ' ' + (n - x - 1) * '_' + s + y * '_' + (m - y) * ' '
            second_line = x * ' ' + '/' + (n - x - 1 + u + y) * ' ' + '\\' + (m - y - 1) * ' '
            if p < q:
                left += [n * ' '] * (q - p)
            elif q < p:
                right += [m * ' '] * (p - q)
            zipped_lines = zip(left, right)
            lines = [first_line, second_line] + [a + u * ' ' + b for a, b in zipped_lines]
            return lines, n + m + u, max(p, q) + 2, n + u // 2

```

```python
class BinaryTree:
    def __init__(self):
        self.root = None

    def getRoot(self):
        return self.root

    def addroot(self,val):
        if self.root is not None:
           print("Tree is not empty!")
        else:
           self.root = Node(val)

    def addleft(self,p, val):
        if (p is None):
            print("Empty Reference!")
            return
        if(p.left is not None):
            print("Left child is already present!")
            return
        else:
            p.left = Node(val)
            return p.left

    def addright(self,p, val):
        if (p is None):
            print("Empty Reference!")
            return
        if (p.right is not None):
            print("Right child is already present!")
            return
        else:
            p.right = Node(val)

            return p.right

    def is_leaf(self, p):
        if (p!=None and p.left==None and p.right==None):
            return True
        else:
            return False

    def deleteTree(self):
        self.root = None

    def inOrder(self):
        if(self.root is not None):
            self._inOrder(self.root)

    def _inOrder(self, node):
        if(node is not None):
            self._inOrder(node.left)
            print(str(node.element) + ' ',end='')
            self._inOrder(node.right)

    def get_leaves(self):
        L = []
        if(self.root is not None):
            self._get_leaves(self.root, L)
        return L

    def _get_leaves(self, node, L):
        if(node is not None):
            self._get_leaves(node.left, L)
            if self.is_leaf(node):
                L.append(node)
            self._get_leaves(node.right, L)

    def get_leaves2(self):
        L = []
        if(self.root is not None):
            self._get_leaves2(self.root, L)
        return L

    def _get_leaves2(self, node, L):
        if(node is not None):
            self._get_leaves2(node.left, L)
            if self.is_leaf(node):
                L.append(node.element)
            self._get_leaves2(node.right, L)



    def subtreecopy(self, node):

        if node is None:
            return node
        else:
            p = Node(node.element)
            p.left = self.subtreecopy(node.left)
            p.right = self.subtreecopy(node.right)
            return p

    def MWrev(self):
        if (self.root is not None):
            return self._MWrev(self.root)
        else: return 0


    def _MWrev(self, node):
        if (node is not None):
            ab = 0
            LA = []
            self._get_leaves2(node.left, LA)
            LB = []
            self._get_leaves2(node.right, LB)
            for x in LA:
                for y in LB:
                    ab += simfunc(x,y)*(N-len(LA)-len(LB))
            la = self._MWrev(node.left)
            lb = self._MWrev(node.right)
            return la + lb + ab

        else:
            return 0



    def __iter__(self):
        if self.root is not None:
            yield from self.root

    def display(self):
        if self.root is not None:
            self.root.display()

```

```python
def parent(obj, node):
    if obj:
        if (obj.left == node ) or (obj.right==node):
            return obj
        else:
            return parent(obj.left, node) or parent(obj.right, node)
```

```python
def sibling(obj,node):
  par = parent(obj , node)
  if par:
    if par.left and par.right:
      if par.left != node:
        return par.left
      else:
        return par.right
```

```python
def count_nodes(node):
    if node is None:
        return 0
    return 1 + count_nodes(node.left) + count_nodes(node.right)
```

```python
def leftsub(tree):
  r = tree.root
  leftree = BinaryTree()
  leftree.root = r.left
  return leftree


def rightsub(tree):
  r = tree.root
  rightree = BinaryTree()
  rightree.root = r.right
  return rightree
```

```python
def Merge(T1, node ,root): ## merge a binary tree with a node

    r = Node(root)
    T = BinaryTree()
    T2 = BinaryTree()
    T2.addroot(node)
    T.root = r
    T.root.left = T1.root
    T.root.right = T2.root

    return T


def merge(T1, T2 ,root): ## merge two binary trees

    r = Node(root)
    T = BinaryTree()
    T.addroot(root)
    T.root.left = T1.root
    T.root.right = T2.root

    return T


def AVGLink(T1,T2): ## compute average link ratio between the leaves of T1 and T2
    L1 = T1.get_leaves()
    L2 = T2.get_leaves()

    ss = 0
    for x in L1:
        for y in L2:
            ss +=  SIM[x,y]

    return ss/(len(L1)*len(L2))

def HAC():  ## average link hierarchical clustering
    LisTree = []
    n = len(SIM)
    for i in range(n):
        b = BinaryTree()
        b.root = Node(i)
        LisTree.append(b)
    while len(LisTree)> 1:
        ### finding best pairs to merge
        bestidx = 0,1
        bestvalue = 0
        for i in range(len(LisTree)):
            for j in range(i):
                p = AVGLink(LisTree[i],LisTree[j])
                if bestvalue < p:
                    bestidx = i,j
                    bestvalue = p

        T = Merge(LisTree[bestidx[0]],LisTree[bestidx[1]])
        LisTree[bestidx[0]] = T
        del LisTree[bestidx[1]]




    return LisTree[0]


def _HAC(LisTree):
    n = len(SIM)
    for i in range(n):
        b = BinaryTree()
        b.root = Node(i)
        LisTree.append(b)
    while len(LisTree)> 1:
        ### finding best pairs to merge
        bestidx = 0,1
        bestvalue = 0
        for i in range(len(LisTree)):
            for j in range(i):
                p = AVGLink(LisTree[i],LisTree[j])
                if bestvalue < p:
                    bestidx = i,j
                    bestvalue = p

        T = Merge(LisTree[bestidx[0]],LisTree[bestidx[1]])
        LisTree[bestidx[0]] = T
        del LisTree[bestidx[1]]

    return LisTree[0]



def wbar(T,x = None):  ## bar{w} function as defined in the OTD paper
    if x == None:
        L = T.get_leaves()
        count  = 0
        ss = 0
        for i in range(len(L)):
            for j in range(i):
                ss = SIM[L[i],L[j]] + ss
                count = count + 1
        return ss/count

    else:
        L = T.get_leaves()
        count = 0
        ss = 0
        for i in range(len(L)):
            ss = SIM[L[i],x] + ss
            count = count  + 1

        return ss/count





def _OTD(T = None ,x = -1):  ## OTD parametric
    if T == None:
        T = BinaryTree()
        T.root = Node(x)
        return T

    elif (T.is_leaf(T.root) or (wbar(T) >= wbar(T,x))):
        S = BinaryTree()
        S.root = Node(x)
        return Merge(T,S)

    else:
        A = BinaryTree()
        A.root = T.root.left
        B = BinaryTree()
        B.root = T.root.right
        if wbar(A,x) <= wbar(B,x):
            return Merge(A,_OTD(B,x))
        else:
            return Merge(_OTD(A,x), B)

def OTD():  ## OTD main
    n = len(SIM)
    idx = list(range(n))
    #np.random.shuffle(idx)

    T = None

    for i in idx:
        T = _OTD(T,i)
    return T

```

```python
def find_depth(root , node, depth=0):
    if node == None:
        return 0
    if root == node:
        return depth

    left_depth = find_depth(root.left, node, depth+1)
    right_depth = find_depth(root.right, node, depth+1)
    return max(left_depth, right_depth)
```

```python
def split(obj,node):
  s=[]
  s.append(node)
  s.append(sibling(obj,node))
  d= find_depth()
  for i in range(2,d+1):
    s.append(sibling(obj,parent(obj,s[i-1])))
  return s
```

```python
def PRCT(T , newpoint ):

     if T == None:
        T = BinaryTree()
        T.addroot(newpoint)
        return T

     else:
        l=T.get_leaves()
        L=[]
        for i in l:
            L.append(i.element)
        root = T.getRoot()
        l = min(L)
        h = max(L)
        v = np.random.uniform(l,h)

        if (T.root.left==None and T.root.right==None) :
           return (Merge(T, newpoint , v ))


        elif all(  np.sign(x-v) != np.sign(newpoint-v)   for x in L  ) :
          return ( Merge(T, newpoint , v ))


        else:
          root = T.getRoot()
          v_ = root.element

          if newpoint <= v_:
            Tmines= leftsub(T)
            Tplus = rightsub(T)
            #newtree = RCT( Tplus , newpoint )
            newtree = merge(PRCT( Tmines , newpoint ) , Tplus , v_ )
            return newtree

          else:
            Tmines = leftsub(T)
            Tplus = rightsub(T)
            newtree = merge( Tmines , PRCT( Tplus , newpoint ), v_ )
            return newtree



def _PRCT():
  g = np.random.normal(0, 1, size=(d, 1))
  c= np.dot(X,g)
  data= c.ravel().tolist()
  T = None
  for i in data:
    T = PRCT(T,i)
  return T
```

```python
def RCT(T , newpoint ):

     if T == None:
        T = BinaryTree()
        T.addroot(newpoint)
        return T

     else:
        l=T.get_leaves()
        L=[]
        for i in l:
            L.append(i.element)
        root = T.getRoot()
        l = np.min(X[:,:d],axis = 0)
        h = np.max(X[:,:d],axis = 0)
        low =[]
        high = []
        for i in range(d):
          low.append(min(l[i],newpoint[i]))
          high.append(max(h[i],newpoint[i]))
        CH =[]
        for i in range(d):
          CH.append(high[i]-low[i])

        if tuple(CH) == (0,)*d:
          j = random.choice(list(range(d)))

        else:
          J = random.choices(list(range(d)), weights= tuple(CH) , k=1)
          j = J[0]

        v = np.random.uniform(l[j],h[j])

        if (T.root.left==None and T.root.right==None) :
           return (Merge(T, newpoint , [v,j] ))


        elif all(  np.sign(x[j]-v) != np.sign(newpoint[j]-v)   for x in L  ) :
          return ( Merge(T, newpoint , [v,j] ))

        else:
          root = T.getRoot()
          v_ = root.element[0]
          j_ = root.element[1]

          if newpoint[j_] <= v_:
            Tmines= leftsub(T)
            Tplus = rightsub(T)
            #newtree = RCT( Tplus , newpoint )
            newtree = merge(RCT( Tmines , newpoint ) , Tplus , [v_,j_] )
            return newtree

          else:
            Tmines = leftsub(T)
            Tplus = rightsub(T)
            newtree = merge( Tmines , RCT( Tplus , newpoint ), [v_,j_] )
            return newtree


def _RCT(M):  ## RCT main
    T = None
    for i in range(M):
        T = RCT(T,X[i])
    return T
```
